# Jumlah Bermain Di UCL Berdasarkan Nama Klub

```python
models = df.groupby('club').count()[['match_played']].sort_values(by='match_played',ascending=True).reset_index()
models = models.rename(columns={'match_played':'numberOfPlayed'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['club'], y=models['numberOfPlayed'], color='royalblue')
plt.xticks(rotation=60)
```

# Distribusi Jumlah Gol

```python
plt.figure(figsize=(15,5))
sns.distplot(df['goals'])
```

# Seleksi Fitur

```python
features = ['right_foot', 'left_foot', 'headers','inside_area', 'outside_areas','penalties']
x = df[features]
y = df['goals']
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

```python
#right_foot:11,left_foot:1,headers:3,inside_area:13,outside_areas:2,penalties:3
input_data = np.array([[11,1,3,13,2,3]])

prediction = lr.predict(input_data)
print('Estimasi jumlah Gol berdasarkan Klub Yaitu  :', prediction,'Goals')
```
